```python
import bs4
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import csv
```

```python
chrome_path = r'C:\Users\Farash\Desktop\chromedriver.exe'
```

```python
driver = webdriver.Chrome(chrome_path)
```

```python
driver.get("https://schoolynetwork.com/ar/")
```

```python
driver.execute_script("var i; while (i < 100000) {document.querySelector('.spNavigationContainer button').click(); i++; if (true) { break ;}}")
```

```python
soup = bs4.BeautifulSoup(driver.page_source , 'lxml')
type(soup)
```

```python
name_list = []
address_list = []
image_list = [] 
elmnhg_list = [] 
y_list = []

for school in soup.find_all('div' , class_= 'restaraCard'):
    name = school.h2.text
    name_list.append(name)
    
    address = school.find('div' , class_ = "restaraRightColumn").text
    address = address.split(":")[1]
    address_list.append(address)

    try:
        image = school.find('div' , class_= "restaraLeftColumn")
        image = image.img['src']
        image_list.append(image)
    except Exception as e :
        image = "School Has No Image!"
        image_list.append(image)
      
    try:
        elmnhg = school.find('div' , class_="spClassViewSelect" )
        elmnhg = elmnhg.find_all('span')[1].text
        elmnhg_list.append(elmnhg)
    except Exception as e :
        elmnhg = "School System Not Defined!"
        elmnhg_list.append(elmnhg)
    
    try:
        years = school.find('ul', class_="spClassMSelect field_school_levels")
        years = years.find_all('li')
        y=[]
        for i in years:
            y.append(i.text)
        y_list.append(y)    
    except Exception as e :
        years = "Education Stages Not Defined!"
        y_list.append(years)
        
        
df = pd.DataFrame({'School Name': name_list, 'Address': address_list , 'Image': image_list , 'Educational System': elmnhg_list , 'Educational Stages': y_list })  
print(df)
df.to_csv(r'D:\github\Dynamic Web Scraping\schools.csv', index=False, encoding='utf_8_sig') #utf_8_sig good with dealing with arabic letters
```
